##ㅁ침adimos KDE al histplot
Las l칤neas que vamos a a침adir son de la librer칤a matplotlib, pero... lo bueno es que podemos combinar ambas librer칤as, y entonces todo es m치s fantas칤a 游댛游뗶

```python
graf = sns.histplot(x=titanic.Age, kde=True) #hemos guardado el gr치fico como una variable
graf.axvline(x=titanic.Age.dropna().mean(), c="red", label="mean")
graf.axvline(x=titanic.Age.dropna().median(), c="green", label="median")
plt.legend()
```

## Boxplot
Este tipo de gr치fico muestra los tres valores del cuartil de la distribuci칩n junto con los valores extremos. Los "bigotes" se extienden a los puntos que se encuentran dentro de 1,5 IQRs del cuartil inferior y superior, y luego las observaciones que caen fuera de este rango se muestran de forma independiente. Esto significa que cada valor en el boxplot corresponde a una observaci칩n real en los datos.

Las l칤neas:  
 - La del medio: la mediana
 - La del medio a la izquierda: Q1: percentil 25%
 - Media de la derecha: Q3: percentil 75%
 
 - Izquierda: Q1 - 1.5 IQR
 - Derecha: Q3 + 1.5 IQR 

Primer cuartil: el 25% de los valores son menores o igual a este valor (punto 2 en el gr치fico anterior).    
Mediana o Segundo Cuartil: Divide en dos partes iguales la distribuci칩n. De forma que  el 50% de los valores son menores o igual a este valor (punto 3 en el gr치fico siguiente).    
Tercer cuartil: el 75% de los valores son menores o igual a este valor (punto 4 en el gr치fico siguiente).    
Rango Intercuart칤lico (RIC): Diferencia entre el valor del tercer cuartil y el primer cuartil.

```python
sns.boxplot(x="Age", data=titanic)
```
